```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("./input/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("./input//test.csv")
test_data.head()
```

```python
#add column senior citizen to data

train_data.loc[train_data['Age'] < 65, 'SeniorCit'] = 0
train_data.loc[train_data['Age'] >= 65, 'SeniorCit'] = 1


test_data.loc[test_data['Age'] < 65, 'SeniorCit'] = 0
test_data.loc[test_data['Age'] >= 65, 'SeniorCit'] = 1
```

```python
train_data.head()
test_data.head()
```

```python
#drop columns
train_data.drop(['Name', 'Ticket'], axis=1, inplace=True)
```

```python
train_data['FOB'] = (train_data['SibSp'] + train_data['Parch'])
test_data['FOB'] = (test_data['SibSp'] + test_data['Parch'])

train_data.head()
```

```python
#drop columns
train_data.drop(['SibSp', 'Parch'], axis=1, inplace=True)
```

```python
train_data.head()
```

```python
# Extract feature and target arrays
X, y = train_data.drop('Survived', axis=1), train_data[['Survived']]
```

```python
features = ["Pclass", "Sex", "Age", "FOB", "Fare", "SeniorCit"]

X = pd.get_dummies(X[features])
X_test = pd.get_dummies(test_data[features])
```

```python
X_test
```

```python
import xgboost as xgb

model = xgb.XGBClassifier(n_estimators=200, objective='binary:logistic', tree_method='hist', eta=0.1, max_depth=3, enable_categorical=True)
```

```python
model.fit(X, y)
```

```python
preds = model.predict(X_test)
```
