```python
dataset.columns
```

```python
offense_df = dataset[['game_id','play_type','yards_gained','offense']]
offense_df.head()
```

```python
#define each plays type
pass_types = ['Pass Reception', 'Passing Touchdown']
rush_types = ['Rush', 'Rushing Touchdown']

#write function to get the play type into the dataframe
def getPlayCall(x):
    if x in pass_types:
            return 'pass'
    elif x in rush_types:
        return 'rush'
    else:
        return 'None'

#get the play type into the dataframe
offense_df['play_call'] = offense_df['play_type'].apply(getPlayCall)


offense_df.head()
```

```python
offense_df = offense_df[offense_df.play_call != 'None']
```

```python
offense_df.head()
```

```python
offense_rush_df = offense_df[['game_id', 'offense', 'play_type','play_call', 'yards_gained']]
offense_rush_df = offense_rush_df[offense_rush_df.play_call != 'pass']
offense_rush_df.head()
```

```python
offense_rush_df = offense_rush_df.groupby(['game_id', 'offense', 'play_call'], as_index=False)['yards_gained'].sum()
offense_rush_df.columns = ['game_id', 'team', 'play_call', 'off_rush_yards']
offense_rush_df
```

```python
offense_pass_df = offense_df[['game_id', 'offense', 'play_type','play_call', 'yards_gained']]
offense_pass_df = offense_pass_df[offense_pass_df.play_call != 'rush']
offense_pass_df.head()
```

```python
offense_pass_df = offense_pass_df.groupby(['game_id', 'offense', 'play_call'], as_index=False)['yards_gained'].sum()
offense_pass_df.columns = ['game_id', 'team', 'play_call', 'off_pass_yards']
offense_pass_df
```

```python
offesnse_game_totals = pd.merge(offense_rush_df[['game_id', 'team', 'off_rush_yards']], offense_pass_df[['game_id', 'team', 'off_pass_yards']], on=['game_id', 'team'], how='outer')
offesnse_game_totals
```
